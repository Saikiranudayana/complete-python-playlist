##### Introduction to Functions
Definition:

A function is a block of code that performs a specific task.
Functions help in organizing code, reusing code, and improving readability.

```python
## syntax
def function_name(parameters):
    """Docstring"""
    # Function body
    return expression
```

```python
#why functions
num = 23
if num%2==0:
    print("the num is even")
else:
    print("the num is  odd")
```

```python
def even_or_odd(num):
    """this function finds even or odd"""
    if num%2==0:
        print("the num is even")
    else:
        print("the num is  odd")
    
```

```python
even_or_odd(9)
```

```python
def print_numbers(*pos):
    for number in pos:
        print(number)
```

```python
print_numbers(1,2,2,2,2,2,"sai","roshan babu")
```

```python
def print_numbers(*args):
    for number in args:
        print(number)
```

```python
def print_deatils(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
        
```

```python
print_deatils(name="roshan babu",age="30",occupations="soilder")
```

```python
def print_deatils(*args ,**kwargs):
    for val in args:
        print(f"postional arguments:{val}")
    
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_deatils(1,1,1,2,2,3,3,4, name="sai",age="30",copuntry="india")
```
