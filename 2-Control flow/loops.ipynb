```python
## while loop

## the while loop excutes as long as the condition is true

count = 0

while count<5:
    print(count)
    count = count+1
   
```

```python
## loop control statements

## break statement

for i in range(1,10):
    if i==6:
        break
    print(i)
    
```

```python
## continue statement

## it will skip the the current iteration and continue with the next iteration of the loop

for i in range(10):
    if i%2==0:
        continue
    print(i)
    
```

```python
#pass 
## the pass satatmet is a null operation it does nothing

for i in range(5):
    if i==3:
        pass
    print(i)
    
```

```python
#nested loops 
# a loop in side a loop

for i in range(5):
    for j in range(4):
        print(f'i={i}, j={j}')
```

```python
## example - calculate the sum of first n natural numbers using a while and for loop


n = int(input('Enter a number: ')) 
total = 0
count = 1


while count<=n:
    total = total+ count
    count = count+1

print (f'the sum of first {n} natural numbers is {total}')
```

```python
n= int(input('Enter a number: ')) 
total = 0 

for i in range(1,n+1):
    total = total+i
    
print (f'the sum of first {n} natural numbers is {total}')
```

```python
# prime numbers between 1 and 100 
for num in range(1,100):
    if num>1:
        for i in range(2,num):
            if num%i == 0 :
                break 
        else:
            print(num)
```
