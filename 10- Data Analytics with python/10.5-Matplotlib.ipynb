#### Data Visualization With Matplotlib

Matplotlib is a powerful plotting library for Python that enables the creation of static, animated, and interactive visualizations. It is widely used for data visualization in data science and analytics. In this lesson, we will cover the basics of Matplotlib, including creating various types of plots and customizing them.

```python
import matplotlib.pyplot as plt 
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('Xaxis')
plt.ylabel('yaxis')
plt.title("Basic line plot")
plt.show()
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]

## customized line plots


plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
```

```python
## Multiple Plots
## Sample data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]


plt.figure(figsize=(9,5))

plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("plot1")

plt.subplot(2,2,2)
plt.plot(y1,x,color='blue')
plt.title("plot2")

plt.subplot(2,2,3)
plt.plot(x,y2,color='red')
plt.title("plot3")

plt.subplot(2,2,4)
plt.plot(y2,x,color='green')
plt.title("plot4")
```

```python
## Bar plots


categories = ['A','B','C','D','E']

values = [5,7,3,8,6]


plt.bar(categories,values,color='purple')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar plot')
plt.show()
```

#### Histograms
Histograms are used to represent the distribution of a dataset. They divide the data into bins and count the number of data points in each bin.

```python
## sample data 
data = [1,2,2,3,3,3,4,4,4,5,5,5,5,5,5]


plt.hist(data,bin=5,color='orange',edgecolor='black')
```

```python
x= [1,2,3,4,5]

y= [4,5,6,7,8]


plt.scatter(x,y,color='blue',marker='x')
```

```python
##pie chart

labels = ['A','B','c','D']
sizes = [30,20,40,10]

colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct="%1.1f%%" ,shadow=True)
```
